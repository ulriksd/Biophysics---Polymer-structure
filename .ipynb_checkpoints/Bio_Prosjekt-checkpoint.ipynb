```python
#Importing libraries

import random
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#Setting parameters

rel_per = 78 # relative permittivity epsilon_r
distance_grid = 23e-6 # distance between two grid points
vacuum_per = 8.85e-12 # Vacuum permittivity
electron = 1.602e-19 # Electron charge
boltzmann = 1.381e-23 # Boltzmanns constant
N = 15
M = 25
```

```python
class Point:
    __slots__ = ("x", "y")

    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __eq__(self, other):
        if not isinstance(other, Point):
            return False
        return self.x == other.x and self.y == other.y

    def __repr__(self):
        return str(self)

    def __str__(self):
        return "("+str(self.x)+","+str(self.y)+")"
    
    def __hash__(self):
        return self.x * 1000 + self.y
```

1a)

```python
def random_pos_sample(N, M):
    grid = np.zeros((N, N)) # defining N x N grid
    sample = random.sample([Point(x, y) for x in range(N) for y in range(N)], 2*M)

    for point in sample[:M]:
        grid[point.y, point.x] = 1.
    for point in sample[M::]:
        grid[point.y, point.x] = -1.
    return grid


def set_grid_monomers(N, M=0):

    grid = random_pos_sample(N, M)

    return grid
```

```python
def grid_scan(grid: np.ndarray):                # Funksjonen gir en liste med alle punkt
    # som er ulik null i grid
    xs, ys = np.nonzero(grid)
    return [Point(int(y), int(x)) for x, y in zip(xs, ys)]


def four_neighbors(N, p: Point):

    if N < 3:
        raise ValueError("Does not have four unique neighbors. N must be at least equal to three")
    return [Point((p.x+1) if p.x + 1 < N else 0, p.y), Point((p.x-1) if p.x > 0 else N-1, p.y), Point(p.x, (p.y+1) if p.y + 1 < N else 0), Point(p.x, (p.y-1) if p.y > 0 else N-1)]
```

1d)

```python
def total_energy(grid):
    N = len(grid)
    
    constant = (electron**2/(4*np.pi*vacuum_per*rel_per*distance_grid**2))
    points = grid_scan(grid)
    energy = constant * sum(grid[point.y, point.x] * sum(grid[nb.x, nb.y] for nb in four_neighbors(N, Point(point.y, point.x))) for point in points)
    return energy * 0.5

print(total_energy(eksempel))
```

1e)

```python
def move(grid, N, M, T):       # used to move polymers and monomers
    beta = 1/(boltzmann*T)
    old_grid = grid.copy()
    E_0 = total_energy(grid)
    points = grid_scan(grid) # opptatte plasser, hvor monomeren ikke kan befinne seg, siden andre befinner seg der
    legal_moves = []
    while len(legal_moves) == 0:

        random_monomer = points[np.random.randint(len(points))] # 2*M = len(points) # velger tilfeldig monomer

        neighbors = four_neighbors(N, random_monomer) # henter nabolaget til monomereren      

        legal_moves = list(set(neighbors) - set(points))

    move = legal_moves[np.random.randint(len(legal_moves))]

    charge = grid[random_monomer.y, random_monomer.x]
    grid[move.y, move.x] = charge
    grid[random_monomer.y, random_monomer.x] = 0.0
    E_1 = total_energy(grid)
    delta_E = E_1-E_0
    if delta_E <= 0 or random.uniform(0,1) <= np.exp(-beta*delta_E):
        return grid, E_1
    else:
        return old_grid, E_1

def MC(N_s, N, M, T, grid):     # T is temperature!
    
    result = grid
    E = np.zeros(N_s)
    n = np.arange(0,N_s,1)
    for i in range(N_s):
        result, E[i] = move(result, N, M, T)
        if i == N_s-1:
            fig, ax = plt.subplots()
            ax.pcolormesh(result[::-1,], cmap="Spectral")
            plt.title(f"blue = positive charge; red = negative charge\n T = {T}")
            ax.set_aspect('equal')
    plt.figure()
    plt.plot(n, E)
    np.savez(f"T={T}.npz", result=result, n=n, E=E)
    plt.show()
```

```python
def find_cluster(p: Point, points, grid, cluster = []):
    N = len(grid)
    neighbors = four_neighbors(N, p)
    for neighbor in neighbors:
        if (neighbor in points) and (neighbor not in cluster):
            cluster.append(neighbor)
            find_cluster(neighbor, points, grid, cluster)
    return cluster

        
def is_in_previous_cluster(point, clusters):
    for cluster in clusters:
        if point in cluster:
            return True
    return False
        

def grid_to_cluster(grid):
    clusters = []
    points = grid_scan(grid)
    dim = len(grid)
    for point in points:
        if is_in_previous_cluster(point, clusters):
            continue
        cluster = [point]
        cluster = find_cluster(point, points, grid, cluster)
        clusters.append(cluster)
    return clusters

def create_cluster_grid(grid, clusters):
    N = len(grid)
    cluster_grid = np.zeros((N, N))
    i = 0
    for cluster in clusters:
        i += 1
        for point in cluster:
            cluster_grid[point.y, point.x] = i
    return cluster_grid

file = np.load(f"T=500.npz")
result500 = file["result"]
cluster_grid500 = create_cluster_grid(result500, grid_to_cluster(result500))
fig, ax = plt.subplots()
ax.pcolormesh(cluster_grid500[::-1,], cmap="bone_r")
ax.set_aspect('equal')


file = np.load(f"T=200.npz")
result200 = file["result"]
cluster_grid200 = create_cluster_grid(result200, grid_to_cluster(result200))
fig, ax = plt.subplots()
ax.pcolormesh(cluster_grid200[::-1,], cmap="bone_r")
ax.set_aspect('equal')
            
```

```python
T_l = 100; T_h = 1000; temperatures = np.linspace(T_l, T_h, 10); N = 15;
M = 25; t_max = 100_000; s = 1/200; C = 10_000;

mean_cluster_size_grid = set_grid_monomers(N, M)

fig, ax = plt.subplots()
ax.pcolormesh(mean_cluster_size_grid[::-1,], cmap="Spectral")
plt.title("blue = positive charge; red = negative charge\n")
ax.set_aspect('equal')

def calculate_mean_cluster_size(cluster_grid, M):
    number_of_clusters = cluster_grid.max()
    mean_cluster_size = 2*M/number_of_clusters
    return mean_cluster_size

def t_equil(T):
    return t_max*np.exp(-s*(T - T_l)) + C

def MC_mean_cluster_size(T):
    t = int(t_equil(T))
    MC(t, N, M, T, mean_cluster_size_grid)
    file = np.load(f"T={T}.npz")
    result = file["result"]
    cluster_grid = create_cluster_grid(result, grid_to_cluster(result))
    mean_cluster_size = calculate_mean_cluster_size(cluster_grid, M)
    return mean_cluster_size

d = np.zeros(len(temperatures))
for i in range(len(temperatures)):    
    d[i] = MC_mean_cluster_size(temperatures[i])

plt.figure()
plt.plot(temperatures, d)
plt.show()
        
    
    
```
